```python
import builtins

user_input = int(builtins.input("type somthing:"))

nature_str = ""
if user_input > 0:
  nature_str = "positive"
elif user_input < 0:
  nature_str = "negative"
else:
  nature_str = "zero"

parity_str = ""
if (user_input % 2) == 0:
  parity_str = "even"
else:
  parity_str = "odd"

print(f"The number {user_input} is {nature_str} and {parity_str}.")
```
