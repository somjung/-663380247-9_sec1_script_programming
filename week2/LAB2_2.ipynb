Activity 2.1: Trace Execution of Conditional Code Snippets


```python
# Snippet A
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"
print(f"Grade: {grade}") # What's the output?

# Snippet B
temp = 25
is_raining = True
if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")
# What's the output?

# Snippet C
age = 17
has_license = True
if age >= 18:
    if has_license:
        print("You can drive legally.")
    else:
        print("You need a license to drive.")
else:
    print("You are too young to drive.")
# What's the output?
```

Snippet A
Answer: เริ่มแรกรับค่า 85 เก็บไว้ที่ score
จากนั้นเช็คเงื่อนไขมากกว่าหรือเท่ากับ 90 ไม่ผ่าน ต่อไปเช็คเงื่อนไข มากกว่าหรือเท่ากับ 80 ผ่าน เก็บ B ที่ grade เมื่อสั่ง print grade จะได้ B

output:Grade: B

Snippet B
Answer: เริ่มแรก เก็บค่า 25 ไว้ที่ temp และ เก็บ true ไว้ที่ is_raining เช็คเงื่อนไขแรก 25 มากกว่า 20 ผ่าน และ is_raining = false ไม่ผ่าน รวมกันด้วยและ ออก มา ไม่ผ่านเงื่อนไขแรก เงื่อนไขต่อมาเช็ค 25 น้อยกว่าหรือเท่ากับ 20 ไม่ผ่าน หรือ is_raining = true ผ่าน รวมกันด้วย หรือ ได้ ผ่าน เข้าเงื่อนไขนี้

output:Stay indoors or bring an umbrella.

Snippet C
Answer: เช็คเงื่อนไขแรก 17 มากกว่าหรือเท่ากับ 18 ไม่ผ่าน เข้าเงื่อนไข else ทันที

output:You are too young to drive.
